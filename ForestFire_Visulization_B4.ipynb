```python
import pandas as pd
import seaborn as sns
from wordcloud import WordCloud
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1.Pie chart
```

```python
sizes = [df['temp'].mean(), df['wind'].mean(), df['rain'].mean()]
explodes = (0.1, 0, 0)
labels = 'temp','wind','rain'
colors = ('red', 'blue', 'yellow')
plt.pie(sizes, explode=explodes, labels=labels, colors=colors, autopct='%1.1f%%',shadow=True, startangle=140)
plt.title('Average Weather')
plt.show()
```

```python
# 2.Bar Plot
```

```python
h = df.iloc[1:21,8]
y_pos = np.arange(len(h))
v = range(1, 21)
plt.bar(y_pos, h, align='center', alpha=0.5)
plt.xticks(y_pos, v)
plt.title('Temperature in last 20 days')
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.show()
```

```python
#3. Histogram Plot
```

```python
e = df.iloc[1:25,8]
plt.hist(e, bins='auto')
plt.xlabel('Temperature')
plt.ylabel('Frequency')
plt.title('Histogram for Temperature Analysis')
plt.show()
```

```python
#4 Line Plot
```

```python
l1 = df.iloc[1:20,8]
l2 = df.iloc[1:20,10]
plt.plot(l1, label='Temperature', color='red', marker='*',linestyle='dotted')
plt.plot(l2, label='wind', color='green', marker='o', linestyle='dashed')
plt.ylim(0,60)
plt.title('Temperature vs Wind')
plt.xlabel('Temp and Wind')
plt.ylabel('values')
plt.legend()
```

```python
#5 Scatter Plot
```

```python
x = df.iloc[1:80,8]
y = df.iloc[1:80,10]
plt.scatter(x,y)
plt.title('Temp vs Wind Scatter plot')
plt.xlabel('temp')
plt.ylabel('Wind')
plt.show()
```
