```python
import pandas as pd
```

```python
df = pd.read_csv(r"G:\B.E\TE VI\DSBDA\DataSets\Heart.csv")
```

```python
#Data Cleaning
```

```python
df.isna().sum()
```

```python
df['Ca'] = df['Ca'].fillna(df['Ca'].mean())
df['Thal'] = df['Thal'].fillna(df['Thal'].mode()[0])
```

```python
df
```

```python
df = df.drop(columns='Unnamed: 0')
```

```python
df
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt
```

```python
# 1.Pie Chart
```

```python
labels = 'RestBP', 'Chol', 'MaxHR'
sizes = [df['RestBP'].mean(), df['Chol'].mean(), df['MaxHR'].mean()]
colors = ['red', 'blue', 'green']
explodes = (0.1, 0, 0)
plt.pie(sizes, explode=explodes,labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Average Report')
```

```python
# 2.Bar Plot
```

```python
import numpy as np
```

```python
h = df.iloc[1:21,4]
y_pos = np.arange(len(h))
v = range(1,21)
plt.bar(y_pos, h, align='center', alpha=0.5)
plt.xticks(y_pos, v)
plt.xlabel('Days')
plt.ylabel('Cholestrol')
plt.title('Cholestrol levels of petient')
plt.show()
```

```python
# 3. Histogram
```

```python
e = df.iloc[:,3]
plt.hist(e, bins='auto')
plt.xlabel('RestBP')
plt.ylabel('Frequency')
plt.title('Histogram for RestBP of Petient')
plt.show()
```

```python
# 4.Line Graph
```

```python
l1 = df.iloc[:16, 3]
l2 = df.iloc[:16, 4]
plt.plot(l1, label='RestBP', color='red', linestyle='dotted', marker='*')
plt.plot(l2, label='Chol', color='blue', linestyle='dashed', marker='o')
plt.ylim(0,470)
plt.legend()
plt.title("RestBp and Cholestrol Comparison") 
plt.ylabel("Value Range")
plt.xlabel("Days") 
```

```python
# 5.Scatter Plot
```

```python
s1 = df.iloc[1:91,0]
s2 = df.iloc[1:91,3]
plt.scatter(s1, s2, color='red')
plt.xlabel('Age')
plt.ylabel('RestBP')
plt.title('Age vs RestBP Relation')
plt.show()
```
