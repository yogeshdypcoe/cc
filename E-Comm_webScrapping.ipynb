```python
request1 = requests.get('https://www.flipkart.com/redmi-note-10-pro-max-dark-night-128-gb/p/itm3b34df1da19de?pid=MOBGB7255D2JWWKM&lid=LSTMOBGB7255D2JWWKMBHBQUP&marketplace=FLIPKART&q=redmi+note+10+pro&store=tyy%2F4io&srno=s_1_1&otracker=search&otracker1=search&fm=organic&iid=16d0b189-833b-46d2-aece-12c9693c2d88.MOBGB7255D2JWWKM.SEARCH&ppt=hp&ppn=homepage&ssid=p8mmdjhx8w0000001713520691423&qH=20ef7d326dcad8f3')
```

```python
soup = bs4.BeautifulSoup(request1.text)
```

```python
soup
```

```python
# Fetch Comments and Reviews
reviews = soup.find_all('div', {'class':'ZmyHeo'})
for review in reviews:
    print(review.getText() , "/n/n")
```

```python
# Fetch Price
price = soup.find('div',{'class':'Nx9bqj CxhGGd'}).getText()
print("Product Price is:-",price)
```

```python
# fetch Product Name
p_name = soup.find('span',{'class':'VU-ZEz'}).getText()
print("Product Name:-",p_name)
```

```python
#Fetch Customer Name
Customer_name = soup.find_all('span',{'class':'BdKevg'})
i=1
print("Customer Names:-")
for cust in Customer_name:
    print(i ,"-",cust.getText())
    i=i+1
```

```python
#Product Ratings
ratings = soup.find('div',{'class':'XQDdHH'}).getText()
print("Overall Ratings:-",ratings)
```

```python
#Individual Ratings
indi_Ratings = soup.find_all('div',{'class':'XQDdHH Ga3i8K'})
for individual_rating in indi_Ratings:
    print("-",individual_rating.getText())
```

```python
hielights = soup.find_all('li',{'class':'_7eSDEz'})
print("Product Hielights:-")
for hie in hielights:
    print("-",hie.getText())
```

```python
prod_desc = soup.find('div',{'class':'yN+eNk w9jEaj'}).getText()
print("\033[1mProduct Description:-\033[0m",prod_desc)
```
